# ***COMPUTER-GUESS-NUMBER***

```python
import random

def guess_the_number_colab():
    print("\nğŸ® Welcome to 'Guess the Number'!")
    print("I'm thinking of a number between 1 and 100.")
    print("Try to guess it in as few attempts as possible!\n")

    number_to_guess = random.randint(1, 100)
    attempts = 0
    guessed_correctly = False

    while not guessed_correctly:
        try:
            guess = int(input("ğŸ”¢ Enter your guess: "))
            attempts += 1

            if guess < number_to_guess:
                print("ğŸ“‰ Too low! Try a higher number.\n")
            elif guess > number_to_guess:
                print("ğŸ“ˆ Too high! Try a lower number.\n")
            else:
                print(f"ğŸ‰ Correct! You guessed the number in {attempts} attempts.\n")
                guessed_correctly = True
        except ValueError:
            print("ğŸš« Invalid input! Please enter a number.\n")

def play_game_colab():
    while True:
        guess_the_number_colab()
        again = input("ğŸ” Do you want to play again? (yes/no): ").strip().lower()
        if again != "yes":
            print("\nğŸ‘‹ Thanks for playing! See you next time.\n")
            break

# Start the game
play_game_colab()
```
