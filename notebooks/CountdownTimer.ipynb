# ***Countdown-Timer***

```python
import time

# Get user input in MM:SS format
user_input = input("â³ Enter time (MM:SS): ")
try:
    minutes, seconds = map(int, user_input.split(":"))
    total_seconds = minutes * 60 + seconds
except ValueError:
    print("âš ï¸ Invalid format. Use MM:SS (e.g., 2:30)")
    exit()

# Countdown loop
while total_seconds > 0:
    mins, secs = divmod(total_seconds, 60)
    current_time = time.strftime("%H:%M:%S")
    print(f"â° Countdown: {mins:02d}:{secs:02d} | ğŸ•’ Clock: {current_time}")
    time.sleep(1)
    total_seconds -= 1

print("\nâœ… Time's up!")
```
