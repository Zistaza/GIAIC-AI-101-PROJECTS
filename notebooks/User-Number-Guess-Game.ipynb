# ***User_NUMBER_Guess-Game***

```python
import random

def computer_guess():
    print("ğŸ§  Think of a number between 1 and 100.")
    input("ğŸ”’ Press Enter when you're ready...")

    low = 1
    high = 100
    attempts = 0

    while True:
        if low > high:
            print("ğŸ˜• Something went wrong. Did you change your number?")
            break

        guess = random.randint(low, high)
        attempts += 1
        print(f"ğŸ¤– My guess is: {guess}")
        print("Type:")
        print(" - 'h' if the guess is too high")
        print(" - 'l' if the guess is too low")
        print(" - 'c' if the guess is correct")

        feedback = input("ğŸ‘‰ Your response (h/l/c): ").strip().lower()

        if feedback == "h":
            high = guess - 1
        elif feedback == "l":
            low = guess + 1
        elif feedback == "c":
            print(f"ğŸ‰ Yay! I guessed your number in {attempts} tries!\n")
            break
        else:
            print("ğŸš« Invalid input! Please enter only h, l, or c.\n")

def play_game():
    while True:
        computer_guess()
        again = input("ğŸ” Do you want to play again? (yes/no): ").strip().lower()
        if again != "yes":
            print("ğŸ‘‹ Thanks for playing!")
            break

# Run the game
play_game()
```
